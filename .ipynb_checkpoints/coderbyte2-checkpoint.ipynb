```python
import requests
from datetime import datetime
import random
import pickle
import json
import os
```

```python
def fetch_data():
    info = requests.get("https://restcountries.com/v3.1/all?fields=name,capital,currencies")
    time = datetime.datetime.now()
    now= time.isoformat()
    info=info.json()
    return info, now
    
```

```python
def fetch_to_disk(binary):
    try:
        print("valor para recibir informacion es", isinstance(binary, bool))
    except TypeError:
        (print("Valor no valido"))
    if binary == True:
        with open("paisinfo.bin", "wb") as bin:
            pickle.dump(fetch_data(), bin)
        os.rename("paisinfo.bin", datetime.now().strftime("%Y-%m-%d %H.%M.%S.bin"))
    elif binary == False:
        with open("paisinfo.json", "w") as file:
            json.dump(fetch_data(), file)
        os.rename("paisinfo.json", datetime.now().strftime("%Y-%m-%d %H.%M.%S.json"))
```
